```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from ipywidgets import interact, FloatSlider, Play
```

$$x' = -y+ax(x^2+y^2)$$
$$y' = x+ay(x^2+y^2)$$

```python
a = -1

def dx_dt(x, a):
    return [-x[1]+a*x[0]*(x[0]**2+x[1]**2), x[0]+a*x[1]*(x[0]**2+x[1]**2)]

Y, X = np.mgrid[-1:1:5j, -1:1:5j]

U, V = dx_dt([X,Y], a)

plt.streamplot(X, Y, U, V, density=0.45, broken_streamlines=False)
plt.grid()
plt.show()
```

```python
y, x = np.mgrid[-1:1:20j, -1:1:20j]
U2, V2 = dx_dt([x,y], a)

Q = plt.quiver(x, y, U2, V2, scale=40)
plt.quiverkey(Q, 1, 1, U=0.5, label='0.5 units')
plt.streamplot(X, Y, U, V, density=0.45, broken_streamlines=False)
plt.show()
```

```python
def plot(a):
    Y, X = np.mgrid[-1:1:5j, -1:1:5j]

    U, V = dx_dt([X,Y], a)

    plt.streamplot(X, Y, U, V, density=0.45, broken_streamlines=False)
    plt.grid()
    plt.show()

interact(plot, a = FloatSlider(value = 1, min = -2, max = 2, step = 0.1))
```

```python
a = 1

def ciclo(x, a):
    return [a*x[0]*(1-(x[0]**2)), 1+x[0]-x[0]]

Y, X = np.mgrid[-5:5:10j, -5:5:10j]

U, V = ciclo([X,Y], a)

plt.streamplot(X, Y, U, V, density=0.4, broken_streamlines=False)
plt.grid()
plt.show()
```

```python
mu = 2

def vander(x, mu):
    return [x[1], (-mu*((x[0]**2)-1)*x[1])-x[0]]

Y, X = np.mgrid[-4:4:10j, -4:4:10j]
U, V = vander([X, Y], mu)

plt.streamplot(X, Y, U, V, density=1, broken_streamlines=False)
plt.grid()
plt.show()
```

```python
def vanderpol(mu):
    
    Y, X = np.mgrid[-4:4:10j, -4:4:10j]
    U, V = vander([X, Y], mu)

    plt.streamplot(X, Y, U, V, density=1, broken_streamlines=False)
    plt.grid()
    plt.show()

interact(vanderpol, mu = FloatSlider(value = 1, min = -2, max = 2, step = 0.1))
```
