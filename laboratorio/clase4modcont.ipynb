```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kepler(t, y):
    return np.array([y[2], y[3], y[0]/(y[0]**2+y[1]**2)**(3/2), y[1]/(y[0]**2+y[1]**2)**(3/2)])
```

```python
def euler_kepler(n,a,b,y0):
    h = (b-a)/n
    t = np.arange(a,b,h)
    y = np.zeros((len(y0), n))
    y[:,0] = y0

    for i in range(n-1):
        y[:,i+1] = y[:,i] + h*(kepler(t[i], y[:,i]))

    return (t, y)
```

```python
y0 = np.array([0.1, 0.2, 0.3, 0.4])
t, y = euler_kepler(1000, 0, 50, y0)
```

```python
plt.plot(t, y[0])
```
