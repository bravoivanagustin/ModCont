```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hutchinson(a,b,n,r,K,tau,N0):
    h = (b-a)/n
    t = np.arange(a,b,h)
    M = [0 for i in range(t.size)]
    Nd = tau/h
    
    for i in range(t.size):
        if t[i]<=0:
            M[i] = N0
    
    for i in range(t.size-1):
        if M[i+1] == 0:
            M[i+1] = M[i] + h*(r*M[i]*(1-(M[int(i-1-Nd)]/K)))

    return (t, M)
```

```python
a = -3
b = 40
n = 1000
r = 1
K = 1
tau = 1
N0 = 0.5
t, N = hutchinson(a,b,n,r,K,tau,N0)
```

```python
plt.plot(t, N)
```

```python
def gravedad(a,b,n,y0,v0,vx):
    h = (b-a)/n 
    t = np.arange(a,b,h)
    x = np.zeros(n)
    y = np.zeros(n)
    v = np.zeros(n)

    y[0] = y0
    v[0] = v0

    for i in range(n-1):
        x[i+1] = vx*t[i+1]
        y[i+1] = y[i] + h*v[i]
        v[i+1] = v[i] + h*(-10)

        if y[i+1] < 0:
            y[i+1] = 0 
            v[i+1] = -v[i+1]*0.8

    
    return (t,x,y,v)
```

```python
t,x,y,v = gravedad(0, 3, 1000, 5, 0, 1)
```

```python
plt.plot(x,y)
```
