```python
import numpy as np
import matplotlib.pyplot as plt
```

***Ejercicio 1***

```python
def temperatura(n, a, b, B, Tr, T0):
    h = (b-a)/n
    t = np.arange(a,b,h)
    T = [0 for i in range(n)]
    T[0] = T0
    
    for i in range(n-1):
        T[i+1] = T[i] +h*B*(T[i]-Tr)

    return (t,T) 
```

```python
t, T = temperatura(1000, 0, 5, np.log(1/4), 20, 36)
```

```python
plt.plot(t,T)
plt.scatter([1,2],[24,21])
plt.plot(t, [4*((1/4)**(i-1))+20 for i in t])
plt.show()
```

***Ejercicio 2***

```python
def glucosa(n, a, b, G, V, k, g0):
    h = (b-a)/n
    t = np.arange(a,b,h)
    g = [0 for i in range(n)]
    g[0] = g0

    for i in range(n-1):
        g[i+1] = g[i] + h*(G/(100*V) - k*g[i])

    return (t,g)
```

```python
G = 1
V = 100
k = 1/100
t, g = glucosa(1000, 0, 1000, G, V, k, 0.001)
```

```python
plt.plot(t, g)
plt.plot(t, [G/(100*V*k) for i in t])
```

***Ejercicio 5***

```python
def poblacion(n, a, b, K, r, A, M0):
    h = (b-a)/n
    t = np.arange(a,b,h)
    M = [0 for i in range(n)]
    M[0] = M0

    for i in range(n-1):
        M[i+1] = M[i] + h*(r*M[i]*((K-M[i])/(K+A*M[i])))

    return (t,M)
```

```python
t, M = poblacion(1000, 0, 100, 100, 1, 15, 1000)
```

```python
plt.plot(t, M)
```
